# Population Visualization on Map


```python

import plotly.express as px

df = px.data.gapminder().query("year==2007")
fig = px.scatter_geo(
    df,
    locations="iso_alpha",
    size="pop",
    hover_name="country",
    color="continent",
    projection="natural earth",
    title="World Population Map"
)
fig.show()
```
