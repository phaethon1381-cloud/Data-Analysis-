# Titanic Analysis


```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic=sns.load_dataset("titanic")
titanic.head()
```

```python
titanic.info()
titanic.describe(include="all")
```

```python
titanic.isnull().sum()
```

```python
sns.countplot(data=titanic,x="class",hue="survived")
plt.show()
```

```python
titanic["sex"].value_counts().plot(kind="bar")
plt.show()
```
